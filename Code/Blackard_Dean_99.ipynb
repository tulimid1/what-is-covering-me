# Recreate figure from Blackard and Dean (1999)

```python
import numpy as np 
import matplotlib.pyplot as plt 
plt.rcParams.update({'font.size': 16})
import seaborn as sns
import pandas as pd
import savingfigR as sf
```

## Figure 4. Comparison of articial neural network and discriminant analysis classification results. 

```python
ann = np.array([70.58, 69.09, 66.47, 63.32, 62.55, 56.94])
lda = np.array([58.38, 58.23, 55.54, 54.97, 49.15, 49.17])
qda = np.array([np.nan, np.nan, np.nan, np.nan, 52.76, 51.06])
num_predictors = np.array([54, 53, 21, 20, 10, 9])
```

```python
df = pd.DataFrame({'Number of Predictors': np.concatenate((num_predictors, num_predictors, num_predictors)),
                    'Model': np.concatenate((np.repeat('NN', 6), np.repeat('LDA', 6), np.repeat('QDA', 6))),
                    'Score': np.concatenate((ann, lda, qda))})
```

```python
plt_styles, bgs = ['seaborn-bright', 'dark_background'], ['_wbg', '_kbg']
for bg, style in zip(bgs, plt_styles):
    with plt.style.context(style):
        fig = plt.figure(figsize=(9,7))
        sns.barplot(x='Number of Predictors', y='Score', hue='Model', palette='colorblind', data=df)
        plt.ylim([0, 100])
        plt.grid()
        plt.show()
        sf.best_save(fig, 'BD_fig4'+bg)
```
